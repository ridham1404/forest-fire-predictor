```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('Algerian_forest_fires_dataset_cleaned_dataset.csv')
```

```python
df.head()
```

```python
df.drop(['day','month','year'],inplace=True,axis=1)
```

```python
df.head()
```

```python
df['Classes'].value_counts()
```

```python
df['Classes']=np.where(df['Classes'].str.contains("not fire"),0,1)
```

```python
df['Classes'].value_counts()
```

```python
#Independent adn dependent features
X=df.drop('FWI',axis=1)
y=df['FWI']
```

```python
X.head()
```

```python
y.head()
```

```python
#TRain Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X_train.shape,X_test.shape
```

## Feature selection Based on Correlation

```python
X_train.corr()
```

```python
#checking for multicolinearity
plt.figure(figsize=(12,10))
corr=X_train.corr()
sns.heatmap(corr,annot=True)
plt.show()
```

```python
def correlation(dataset,thresold):
     col_corr=set()
     corr_matrix=dataset.corr()
     for i in range(len(corr_matrix.columns)):
          for j in range(i):
               if abs(corr_matrix.iloc[i,j])>thresold:
                    col_name=corr_matrix.columns[i]
                    col_corr.add(col_name)
     return col_corr               
```

```python
## Thresold valus is decided by domain experts
corr_features=correlation(X_train,0.85)
```

```python
#drop features when correlation is more than 0.85
X_train.drop(corr_features,inplace=True,axis=1)
X_test.drop(corr_features,inplace=True,axis=1)
```

```python
X_train.shape,X_test.shape
```

```python
#Feature Scalling or Standardization
from sklearn.preprocessing import StandardScaler 
```

```python
scaler=StandardScaler()
```

```python
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
X_train_scaled
```

## Box plot to understand Effect of Standard Scaler


```python
plt.subplots(figsize=(15,5))
plt.subplot(1,2,1)
sns.boxplot(data=X_train)
plt.title('X_train Before Scaling')
plt.subplot(1,2,2)
sns.boxplot(data=X_train_scaled)
plt.title('X_train After Scaling')
plt.show()
```

## Model Training

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
linereg=LinearRegression()
linereg.fit(X_train_scaled,y_train)
y_pred=linereg.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Maen Absolute Error : ",mae)
print("R2 Score : ",score)
plt.scatter(y_test,y_pred)
plt.show()
```

## Lasso Regression model

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
lasso=Lasso()
lasso.fit(X_train_scaled,y_train)
y_pred=lasso.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Maen Absolute Error : ",mae)
print("R2 Score : ",score)
plt.scatter(y_test,y_pred)
plt.show()
```

## Ridge Regression Model

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
ridge=Ridge()
ridge.fit(X_train_scaled,y_train)
y_pred=ridge.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Maen Absolute Error : ",mae)
print("R2 Score : ",score)
plt.scatter(y_test,y_pred)
plt.show()
```

## ElasicNet Regression Model


```python
from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
elastic=ElasticNet()
elastic.fit(X_train_scaled,y_train)
y_pred=elastic.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Maen Absolute Error : ",mae)
print("R2 Score : ",score)
plt.scatter(y_test,y_pred)
plt.show()
```
